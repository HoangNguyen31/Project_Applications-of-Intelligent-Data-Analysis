## **2.6. Với mỗi cột số, các giá trị được phân bố như thế nào?**

```python
numeric_df = df.select_dtypes(include=["number"])
nume_col_df = numeric_df.describe()
nume_col_df.loc["missing_count"] = numeric_df.isna().sum()
nume_col_df.loc["missing_ratio"] = (numeric_df.isna().sum() / numeric_df.shape[0]) * 100
nume_col_df
```

Nhận xét:
- Các cột `category (Mã bất động sản)` và `region (Mã tỉnh)` đều chỉ có 1 giá trị duy nhất lần lượt là `1010` và `13`. Ta cần cân nhắc loại bỏ 2 cột này.
- Các cột thuộc tính còn lại có một số giá trị bị thiếu, đặc biệt là ở các cột như `ward`, `toilets`, `longitude`, `latitude`, và `deposit`. Cần xem xét các biện pháp để xử lý dữ liệu bị thiếu này.
- Các cột như `deposit`, `price_million_per_m2`, `price` đòi hỏi kiểm tra kỹ lưỡng để loại bỏ các giá trị ngoại lai.

## **2.7. Với mỗi cột phân loại, các giá trị được phân bổ như thế nào?**

```python
object_df = df.select_dtypes(include="object")
object_col_df = object_df.describe()
object_col_df.loc["missing_count"] = object_df.isna().sum()
object_col_df.loc["missing_ratio"] = (object_df.isna().sum() / object_df.shape[0]) * 100
object_col_df
```

Nhận xét:

- Các cột thuộc tính còn lại có một số giá trị bị thiếu. Cần xem xét các biện pháp để xử lý dữ liệu bị thiếu này

## **2.8. Loại bỏ các cột thuộc tính không có ý nghĩa**

```python
column_to_drop = ["category", "region", "category_name", "region_name", "type", "type_name"]
df.drop(column_to_drop, axis=1, inplace=True)
```

## **2.8. Tình trạng khuyết dữ liệu**

```python
null_df = pd.DataFrame(df.isnull().sum(), columns=["Số mẫu bị thiếu"])
null_df["Tỉ lệ bị thiếu"] = np.round(df.isnull().sum() / df.shape[0] * 100, 2)
null_df.index.name = "Thuộc tính"
null_df = null_df.sort_values(by="Tỉ lệ bị thiếu", ascending=False)
null_df.loc[null_df["Tỉ lệ bị thiếu"] > 0]
```

Nhận xét:
- Thuộc tính `deposit` là thuộc tính thiếu nhiều nhất, khoảng `58.20%` dữ liệu bị thiếu.
- Các thuộc tính thiếu còn lại có số lượng dữ liệu bị thiếu không đáng kể.
