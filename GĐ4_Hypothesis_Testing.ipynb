## **4.2. Kiểm định giả thuyết**

### **4.2.1: Giá thuê trung bình căn hộ, chung cư ở thành phố Thủ Đức có cao hơn so với giá thuê trung bình căn hộ, chung cư  trên toàn bộ thành phố Hồ Chí Minh?**
- Giả thuyết null (H0): Giá thuê trung bình nhà ở thành phố Thủ Đức không cao hơn so với thành phố Hồ Chí Minh.

  H0: $\mu \leq \mu_0$

- Giả thuyết thay thế (H1): Giá thuê trung bình nhà ở thành phố Thủ Đức cao hơn so với thành phố Hồ Chí Minh.

  H1: $\mu > \mu_0$

```python
TD_area_price = df.loc[df["area_name"] == "Thành phố Thủ Đức"]["price"]
HCM_price = df["price"]

t_statistic, p_value = stats.ttest_ind(TD_area_price, HCM_price, alternative='greater')

print("Giá trị t-statistic:", t_statistic)
print("Giá trị p-value:", p_value)
```

```python
alpha = 0.05
if p_value < alpha:
    print("Có đủ bằng chứng để bác bỏ giả thiết null (H0).")
    print("Giá thuê trung bình nhà ở thành phố Thủ Đức cao hơn so với thành phố Hồ Chí Minh.")
else:
    print("Không đủ bằng chứng để bác bỏ giả thiết null (H0).")
    print("Không có sự khác biệt đáng kể giữa giá thuê trung bình nhà ở thành phố Thủ Đức cao hơn so với thành phố Hồ Chí Minh.")
```

### **4.2.2. Số phòng ngủ của Căn hộ dịch vụ, mini bằng với số phòng ngủ của Chung cư**
- Giả thuyết null (H0): Số phòng ngủ của Căn hộ dịch vụ, mini bằng với số phòng ngủ của Chung cư.

  H0: $\mu = \mu_0$

- Giả thuyết thay thế (H1): Số phòng ngủ của Căn hộ dịch vụ, mini khác với số phòng ngủ của Chung cư.

  H1: $\mu \neq \mu_0$

```python
type_1 = df.loc[df["apartment_type"] == "Căn hộ dịch vụ, mini"]["rooms"]
type_2 = df.loc[df["apartment_type"] == "Chung cư"]["rooms"]

t_statistic, p_value = stats.ttest_ind(type_1, type_2)

print("Giá trị t-statistic:", t_statistic)
print("Giá trị p-value:", p_value)
```

```python
alpha = 0.05
if p_value < alpha:
    print("Có đủ bằng chứng để bác bỏ giả thiết null (H0).")
    print("Số phòng ngủ của Căn hộ dịch vụ, mini khác với số phòng ngủ của Chung cư.")
else:
    print("Không đủ bằng chứng để bác bỏ giả thiết null (H0).")
    print("Số phòng ngủ của Căn hộ dịch vụ, mini bằng với số phòng ngủ của Chung cư.")
```

### **4.2.3: Có sự khác biệt giữa các quận về giá thuê căn hộ, chung cư?**
- Giả thuyết null (H0): Không có sự khác biệt đáng kể về giá thuê căn hộ, chung cư giữa các quận.
- Giả thuyết thay thế (H1): Có ít nhất một sự khác biệt đáng kể về giá thuê căn hộ, chung cư giữa ít nhất hai quận.

```python
f_statistic, p_value = stats.f_oneway(*[group["price"].values for name, group in df.groupby("area_name")])

print("F-statistic:", f_statistic)
print("P-value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
  print("Có đủ bằng chứng để bác bỏ giả thiết null (H0).")
  print("Có ít nhất một sự khác biệt đáng kể về giá thuê căn hộ, chung cư giữa ít nhất hai quận.")
else:
  print("Không đủ bằng chứng để bác bỏ giả thiết null (H0).")
  print("Không có sự khác biệt đáng kể về giá thuê căn hộ, chung cư giữa các quận.")
```

### **4.2.4: Có sự khác biệt giữa các loại căn hộ, chung cư về giá thuê căn hộ, chung cư?**
- Giả thuyết null (H0): Không có sự khác biệt đáng kể về giá thuê căn hộ, chung cư giữa các loại căn hộ, chung cư.
- Giả thuyết thay thế (H1): Có ít nhất một sự khác biệt đáng kể về giá thuê căn hộ, chung cư giữa ít nhất hai loại căn hộ, chung cư.

```python
f_statistic, p_value = stats.f_oneway(*[group["price"].values for name, group in df.groupby("apartment_type")])

print("F-statistic:", f_statistic)
print("P-value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
  print("Có đủ bằng chứng để bác bỏ giả thiết null (H0).")
  print("Có ít nhất một sự khác biệt đáng kể về giá thuê căn hộ, chung cư giữa ít nhất hai loại căn hộ, chung cư.")
else:
  print("Không đủ bằng chứng để bác bỏ giả thiết null (H0).")
  print("Không có sự khác biệt đáng kể về giá thuê căn hộ, chung cư giữa các loại căn hộ, chung cư.")
```

### **4.2.5: Liệu số phòng ngủ độc lập hay phụ thuộc vào loại căn hộ, chung cư?**
- Giả thuyết null (H0): Số phòng ngủ độc lập với loại căn hộ, chung cư.
- Giả thuyết thay thế (H1): Số phòng ngủ phụ thuộc vào loại căn hộ, chung cư.

```python
contingency_table = pd.crosstab(df["rooms"], df["apartment_type"])
chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)
print("Giá trị thống kê chi-square:", chi2)
print("p-value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
  print("Có đủ bằng chứng để bác bỏ giả thiết null (H0).")
  print("Số phòng ngủ độc lập với loại căn hộ, chung cư.")
else:
  print("Không đủ bằng chứng để bác bỏ giả thiết null (H0).")
  print("Số phòng ngủ phụ thuộc vào loại căn hộ, chung cư.")
```
